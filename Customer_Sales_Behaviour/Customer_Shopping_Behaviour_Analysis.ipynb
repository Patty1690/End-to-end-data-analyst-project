## Handling null values

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

## Replacing Column Names

```python
df.columns= df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

## Creating a column for Age_Group

```python
labels = ['Young','Adult','Middle_Age','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head(5)
```

```python
df.head()
```

## Creating a column for purchases_frequency_days

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_map = {
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}
```

```python
df['frequency_of_purchases'] = df['frequency_of_purchases'].map(frequency_map)
```

```python
df[['frequency_of_purchases','frequency_of_purchases']].head(10)
```

```python
df.head()
```
